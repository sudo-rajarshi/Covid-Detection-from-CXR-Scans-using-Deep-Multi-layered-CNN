```python
import pandas as pd
import os
import shutil
import random
```

```python
csv_path = '/home/rajarshi/Documents/DATASETS/Health/COVID/covid-chestxray-dataset-master/metadata.csv'
covid_image_path = '/home/rajarshi/Documents/DATASETS/Health/COVID/covid-chestxray-dataset-master/images'
normal_image_path = '/home/rajarshi/Documents/DATASETS/Health/COVID/pneumonia/chest_xray/test/NORMAL'
```

## COVID

```python
df = pd.read_csv(csv_path)
```

```python
df.shape
```

```python
dataset = '/home/rajarshi/Documents/DATASETS/Health/COVID/covid-chestxray-dataset-master/DATASET'
train = os.path.join(dataset, 'train')
test = os.path.join(dataset, 'test')
covid = os.path.join(dataset, train, 'covid')
normal = os.path.join(dataset, train, 'normal')

if not os.path.exists(dataset):
    os.mkdir(dataset)
else:
    shutil.rmtree(dataset)
    os.mkdir(dataset)

os.mkdir(train)
os.mkdir(test)
os.mkdir(covid)
os.mkdir(normal)
```

```python
for (i, row) in df.iterrows():
    print(row['view'])
```

```python
cnt = 0
for (i, row) in df.iterrows():
    if row['finding'] == 'COVID-19' and row['view'] == 'PA':
        cnt +=1
print(cnt)
```

```python
cnt_covid = 0
for (i, row) in df.iterrows():
    if row['finding'] == 'COVID-19' and row['view'] == 'PA':
        filename = row['filename']
        src_covid = os.path.join(covid_image_path, filename)
        dst_covid = os.path.join(covid, filename)
        shutil.copy2(src_covid, dst_covid)
        cnt_covid += 1
        
print(cnt_covid)
```

## Normal

```python
normal_ls = []
cnt_normal = 0
for image_name in os.listdir(normal_image_path):
    if cnt_normal == cnt_covid:
        break
    else:
        normal_ls.append(image_name)
        cnt_normal += 1
```

```python
len(normal_ls)
```

```python
normal_ls[0]
```

```python
for image_name in normal_ls:
    src_normal = os.path.join(normal_image_path, image_name)
    dst_normal = os.path.join(normal, image_name)
    shutil.copy2(src_normal, dst_normal)
```
